# **Fashion Mnist Project Using Deep Neural Network**

**Import Libraries**

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Load Data**

```python
(x_train, y_train), (x_test, y_test) =tf.keras.datasets.fashion_mnist.load_data()
```

**Check Shape**

```python
x_test.shape, x_train.shape, y_test.shape, y_train.shape
```

```python
x_train[0]
```

**Show One Image**

```python
plt.imshow(x_train[0], cmap ="Greys")
```

**Show Images**

```python
class_labels = ["T-shirt/top","Trouser","Pullover","Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankle boot"]

plt.figure(figsize=(16,16))
for i in range(25):
  plt.subplot(5,5,i+1)
  plt.imshow(x_train[i],cmap="Greys")
  plt.axis('off')
  plt.title(class_labels[y_train[i]]+"="+str(y_train[i]), fontsize=20)
```

**Feature Scaling**

```python
x_train = x_train/255.
x_test = x_test/255.
```

```python
x_train[0]
```

# **Build Deep Neural Network Model**

```python
model = tf.keras.Sequential([
                             tf.keras.layers.Flatten(input_shape = (28,28)),
                             tf.keras.layers.Dense(32,activation='relu'),
                             tf.keras.layers.Dense(32, activation='relu'),
                             tf.keras.layers.Dense(10,activation='softmax')
])
```

**Compile The Model**

```python
model.compile(loss = tf.keras.losses.sparse_categorical_crossentropy,
              optimizer = tf.keras.optimizers.Adam(),
              metrics=['accuracy'])
```

**Fit The Model**

```python
model.fit(x_train,y_train, epochs=100)
```

**Evaluate and Test The Model**

```python
model.evaluate(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
```

```python
np.argmax(y_pred[4])
```

```python
y_test[4]
```

**Prediction Images & Actual Images**

```python
plt.figure(figsize=(16,16))
for i in range(25):
  plt.subplot(5,5,i+1)
  plt.imshow(x_test[i],cmap='Blues')
  plt.axis('off')
  plt.title("Actual= {} \n Predicted = {}".format(class_labels[y_test[i]], class_labels[np.argmax(y_pred[i])]))
```

# **Create Confusion Matrix**

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, [ np.argmax(i) for i in y_pred])
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(cm, annot=True, fmt = "d")
```

**Classification Report**

```python
from sklearn.metrics import classification_report
cr = classification_report(y_test, [ np.argmax(i) for i in y_pred], target_names = class_labels,)
```

```python
print(cr)
```

**Save The Model**

```python
model.save('Fashion_Mnist_Project_Using_Deep_Neural_Network')
```

**Load The Model**

```python
loaded_model = tf.keras.models.load_model('Fashion_Mnist_Project_Using_Deep_Neural_Network')
```
